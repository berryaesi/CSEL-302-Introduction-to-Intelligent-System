# **Key Findings**

***Product Type and Price***

```python
# Scatter Plots
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Price', y='Revenue generated', hue='Product type', size='Number of products sold')
plt.title("Scatter Plot of Price vs. Revenue generated")
plt.show()
```

***Sales by Product Type***

```python
# Calculate total sales by product type
sales_data = data.groupby('Product type')['Number of products sold'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales by Product Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The company derives more revenue from skincare products, and the higher the price of skincare products, the more revenue they generate.
So 45% of the business comes from skincare products, 29.5% from haircare, and 25.5% from cosmetics.

***Total Revenue by Shipping Carrier***

```python
# Group data by 'Shipping carriers' and sum 'Revenue generated'
total_revenue = data.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()

# Create a bar plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(total_revenue['Shipping carriers'], total_revenue['Revenue generated'])
plt.title('Total Revenue by Shipping Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Revenue Generated')
plt.xticks(rotation=45)
plt.show()
```

In one of the above visualizations, we discovered that Carrier B helps the company in more revenue. It is also the most costly Carrier among the three.

***Product type***

```python
avg_lead_time = data.groupby('Product type')['Lead time'].mean().reset_index()
avg_manufacturing_costs = data.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time, avg_manufacturing_costs, on='Product type')
result.rename(columns={'Lead time': 'Average Lead Time', 'Manufacturing costs': 'Average Manufacturing Costs'}, inplace=True)
print(result)
```

***Revenue generated by SKU***

```python
# Sort the data by SKU for better visualization
sorted_data = data.sort_values(by='SKU')

# Create a line plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(sorted_data['SKU'], sorted_data['Revenue generated'], marker='o', linestyle='-')
plt.title('Revenue Generated by SKU')
plt.xlabel('SKU')
plt.ylabel('Revenue generated')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

***Stock Levels by SKU***

```python
# Create a line plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(data['SKU'], data['Stock levels'], marker='o', linestyle='-')
plt.title('Stock Levels by SKU')
plt.xlabel('SKU')
plt.ylabel('Stock levels')
plt.grid(True)
plt.show()
```

***Order Quantity by SKU***

```python
# Create bar chart using Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(data['SKU'], data['Order quantities'])
plt.xlabel('SKU')
plt.ylabel('Order Quantities')
plt.title('Order Quantity by SKU')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***Shipping Costs by Carrier***

```python
# Group the data by 'Shipping carriers' and calculate the mean 'Shipping costs' for each carrier
grouped_data = data.groupby('Shipping carriers')['Shipping costs'].mean()

# Convert the grouped data into lists for Matplotlib
carriers = grouped_data.index.tolist()
costs = grouped_data.tolist()

# Create the bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(carriers, costs)
plt.xlabel('Shipping carriers')
plt.ylabel('Shipping costs')
plt.title('Shipping Costs by Carrier')
plt.xticks(rotation=45)
plt.show()
```

***Cost Distribution by Transportation Mode***

```python
# Count plot for transportation modes
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Transportation modes')
plt.title('Cost Distribution by Transportation Mode')
plt.xlabel('Transportation Mode')
plt.ylabel('Count')
plt.show()
```

So the company spends more on Road and Rail modes of transportation for the transportation of Goods

# **Advance Analysis**

***Average Defect Rates by Product Type***

```python
# Group data by 'Product type' and calculate mean 'Defect rates'
defect_rates_by_product = data.groupby('Product type')['Defect rates'].mean().reset_index()

# Create bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=defect_rates_by_product, x='Product type', y='Defect rates')
plt.title('Average Defect Rates by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Defect Rates')
plt.xticks(rotation=45)
plt.show()
```

So the defect rate of haircare products is higher.

***Total Revenue Generated From Each States***

```python

# Aggregate data
total_revenue = data.groupby('Location')['Revenue generated'].sum().reset_index()

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=total_revenue, x='Location', y='Revenue generated')
plt.title('Total Revenue based on location')
plt.xlabel('Locations')
plt.ylabel('Revenue generated')
plt.xticks(rotation=45)
plt.show()
```

***Defect Rates by Transportation Mode***

```python
# Create a pivot table
pivot_table = pd.pivot_table(data, values='Defect rates',
                             index=['Transportation modes'],
                             aggfunc='mean')

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=pivot_table.index, y='Defect rates', data=pivot_table, palette='pastel')
plt.title('Defect Rates by Transportation Mode')
plt.xlabel('Transportation Mode')
plt.ylabel('Mean Defect Rate')
plt.xticks(rotation=45)
plt.show()
```

Road transportation results in a higher defect rate, and Air transportation has the lowest defect rate.
